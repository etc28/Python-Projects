```python
#importing libraries to use
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
import matplotlib.pyplot as plt
import numpy as np
```

```python
#connecting to url 
url = "https://en.wikipedia.org/wiki/List_of_most-polluted_cities_by_particulate_matter_concentration"

r = requests.get(url)
```

```python
#using beautiful soup to read the webpage 
soup = BeautifulSoup(r.content, 'html5lib')

table = soup.findAll('table', attrs = {'class': 'wikitable sortable'}) #reading the table removes headers and text from page
```

```python
#reading the text from the table
for row in table:
    print(row.text)
```

```python
#creating a list 
list_rows = []
for r in row:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '', str_cells))
    list_rows.append(clean2)

df = pd.DataFrame(list_rows)

df.head(10)
```

```python
df1 = df[0].str.split(',', expand=True)
df1.head(10)
```

```python
df1[0] = df1[0].str.strip('[]')
df1.head(10)
```

```python
col_labels = soup.find_all('th')

all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
df3 = df2[0].str.split(',', expand=True)
df3.head()
```

```python
frames = [df3, df1]
df4 = pd.concat(frames)
df4.head(10)
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
#Since removing NA will remove most of the columns I am filling in with 0 instead
df6 = df5.fillna(0)
df6
```

```python
#removing top empty row
df7 = df6.drop(df6.index[0])
df7.head(10)
```

```python
#looking for duplicates
df7.duplicated()
```

```python
#removing any duplicates
df7.drop_duplicates()
```
